```python
# Import library pymongo
from pymongo import MongoClient

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
collection = db["employees"]

# ======== INSERT DATA ========
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}
collection.insert_one(employee_data)
print('Data Alice inserted')

# Insert beberapa data tambahan
employees = [
    {'name': 'Ahmada', 'department': 'IT', 'age': 21, 'salary': 50000},
    {'name': 'Nopal', 'department': 'HR', 'age': 20, 'salary': 40000},
    {'name': 'Ilham', 'department': 'Marketing', 'age': 35, 'salary': 60000}
]
collection.insert_many(employees)
print('Beberapa data tambahan berhasil diinsert')

# ======== UPDATE DATA ========
# Misal: Update gaji Bob menjadi 6500
collection.update_one({'name': 'Ahmada'}, {'$set': {'salary': 65000}})
print("Data Ahmada berhasil diupdate")

# ======== DELETE DATA ========
# Misal: Hapus data karyawan bernama Charlie
collection.delete_one({'name': 'Ilham'})
print("Data Ilham berhasil dihapus")

# ======== QUERY DATA ========
print("\nData karyawan saat ini:")
for employee in collection.find():
    print(employee)
```

```python
# Import library
from pymongo import MongoClient

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
collection = db["employees"]

# ====== AGGREGATION PIPELINE ======
# Menghitung rata-rata gaji per departemen
pipeline = [
    {
        '$group': {
            '_id': '$department',  # Group berdasarkan departemen
            'average_salary': {'$avg': '$salary'}  # Hitung rata-rata gaji
        }
    },
    {
        '$sort': {'average_salary': -1}  # (opsional) urutkan dari gaji tertinggi
    }
]

# Jalankan aggregation dan tampilkan hasilnya
print("Rata-rata gaji per departemen:\n")
for result in collection.aggregate(pipeline):
    print(f"Departemen: {result['_id']}, Rata-rata Gaji: {result['average_salary']:.2f}")
```

```python
# 1Ô∏è‚É£ Hitung rata-rata harga semua produk
avg_price_result = products_collection.aggregate([
    {"$group": {"_id": None, "avg_price": {"$avg": "$price"}}}
])

# Ambil nilai rata-rata dari hasil agregasi
avg_price = list(avg_price_result)[0]["avg_price"]

print(f"üí∞ Rata-rata harga produk: {avg_price:,.0f}")

# 2Ô∏è‚É£ Query untuk menemukan produk dengan harga di atas rata-rata
expensive_products = products_collection.find({"price": {"$gt": avg_price}})

print("\nüì¶ Produk dengan harga di atas rata-rata:")
for product in expensive_products:
    print(f"- {product['name']} ({product['price']:,.0f})")
```

```python
# Pipeline untuk menghitung total produk per kategori
pipeline = [
    {
        "$group": {
            "_id": "$category",          # Kelompokkan berdasarkan kategori
            "total_products": {"$sum": 1}  # Hitung jumlah produk di setiap kategori
        }
    },
    {
        "$sort": {"total_products": -1}   # (Opsional) Urutkan dari kategori dengan produk terbanyak
    }
]

# Jalankan pipeline
results = products_collection.aggregate(pipeline)

# Tampilkan hasil
print("üìä Total produk per kategori:")
for result in results:
    print(f"- {result['_id']}: {result['total_products']} produk")
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
karyawan_collection = db["karyawan"]

# Aggregation untuk mencari 5 karyawan dengan gaji tertinggi di setiap departemen
pipeline = [
    {"$sort": {"departement": 1, "salary": -1}},  # Urutkan berdasarkan departemen dan gaji tertinggi
    {"$group": {
        "_id": "$departement",
        "top_karyawan": {"$push": {"name": "$name", "salary": "$salary", "age": "$age"}},
    }},
    {"$project": {
        "_id": 0,
        "departement": "$_id",
        "top_5": {"$slice": ["$top_karyawan", 5]}  # Ambil hanya 5 karyawan teratas
    }}
]

results = list(karyawan_collection.aggregate(pipeline))

# Tampilkan hasilnya
for dept in results:
    print(f"\nDepartemen: {dept['departement']}")
    for i, k in enumerate(dept['top_5'], start=1):
        print(f"{i}. {k['name']} - Gaji: {k['salary']} - Umur: {k['age']}")
```

```python
from pymongo import MongoClient

# Koneksi ke database
client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
karyawan_collection = db["karyawan"]

# Aggregation: total gaji & rata-rata umur per departemen
pipeline = [
    {
        "$group": {
            "_id": "$departement",
            "total_gaji": {"$sum": "$salary"},
            "rata_rata_umur": {"$avg": "$age"},
            "jumlah_karyawan": {"$sum": 1}
        }
    },
    {
        "$project": {
            "_id": 0,
            "departement": "$_id",
            "total_gaji": 1,
            "rata_rata_umur": {"$round": ["$rata_rata_umur", 2]},
            "jumlah_karyawan": 1
        }
    },
    {"$sort": {"departement": 1}}  # urutkan alfabetis
]

results = list(karyawan_collection.aggregate(pipeline))

# Tampilkan hasil laporan
print("=== Laporan Gaji dan Umur Karyawan per Departemen ===\n")
for dept in results:
    print(f"Departemen       : {dept['departement']}")
    print(f"Total Gaji       : Rp {dept['total_gaji']:,}")
    print(f"Rata-rata Umur   : {dept['rata_rata_umur']} tahun")
    print(f"Jumlah Karyawan  : {dept['jumlah_karyawan']}\n")
```
